```python
names = ['*gridT*', '*gridU*', '*gridV*', '*gridW*', '*icemod*']

for year in range(2002, 2014):
    L=365
    if (year%4==0):
        L=366
    base1 = datetime.datetime(year, 12, 31)
    date_list1 = [base1 - datetime.timedelta(days=x) for x in range(0, L, 5)]
    #
    base2 = datetime.datetime(year, 12, 27)
    date_list2 = [base2 - datetime.timedelta(days=x) for x in range(0, L, 5)]
    for i in range(len(date_list1)):
        str1=date_list1[i].strftime('y%Ym%md%d*')
        str2=date_list2[i].strftime('y%Ym%md%d*')
        for j in range(5):
            print('mv'+' '+names[j]+str1+' '+names[j]+str2)
```
