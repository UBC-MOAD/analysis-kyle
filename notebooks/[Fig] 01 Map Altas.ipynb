```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
fig=plt.figure(figsize=(20, 20))
proj = Basemap(llcrnrlon=-135.,llcrnrlat=45,urcrnrlon=-120.,urcrnrlat=55,\
            resolution='f', projection='lcc',\
            lat_1=30, lat_2=60, lat_0=45, lon_0=-120)
proj.drawcoastlines()
#proj.fillcontinents(color='coral',lake_color='aqua')
# draw parallels and meridians.
proj.drawparallels(np.arange(30, 60, 2.5))
proj.drawmeridians(np.arange(-150, -120, 2.5))
#proj.drawmapboundary(fill_color='aqua')
plt.title("Lambert Conformal Projection", fontsize=14, fontweight='bold')
plt.show()
```
