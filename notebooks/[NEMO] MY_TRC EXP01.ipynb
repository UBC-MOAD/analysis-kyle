```python
#import copy
import glob
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
from __future__ import division
from __future__ import print_function
from mpl_toolkits.basemap import cm
from mpl_toolkits.basemap import Basemap
from mpl_toolkits.basemap import shiftgrid
% matplotlib inline
```

```python
ptrc_filename=glob.glob('_data/EXP01*.nc')
ptrc_obj = nc.Dataset(ptrc_filename[0])
```

```python
ptrc_obj.variables.keys()
```

```python
Ba=ptrc_obj.variables['Ba'][:]
latitude=ptrc_obj.variables['nav_lat'][:]
longitude=ptrc_obj.variables['nav_lon'][:]
```

```python
Ba.shape
```

```python
plt.hist(Ba.flatten(), [0, 1e7]);
```

```python
fig=plt.figure(figsize=(15, 15)); ax=plt.gca()
proj=Basemap(projection='npstere', resolution='c', boundinglat=50, lon_0=90, round=True, ax=ax)
proj.drawparallels(np.arange(-90, 90, 15), labels=[1, 1, 1, 1], fontsize=10, latmax=90)
proj.drawmeridians(np.arange(0, 360, 60), labels=[1, 1, 1, 1], fontsize=10, latmax=90)
proj.drawcoastlines(linewidth=2.0, linestyle='-', color='k')
x, y=proj(longitude, latitude)
CS=proj.pcolor(x, y, Ba[0, 0, :, :], vmin=0, vmax=1, cmap=plt.cm.hot_r)
ax.set_title('Modelled Barium concentration (Jan)', fontsize=14, fontweight='bold', y=1.025)
```
